```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('train.csv')

df = df_original.copy()
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(x="area_sqm", y="price", data=df, hue='type', hue_order=df['type'].unique().sort())
```

```python
flat_model_order = sorted(df['flat_model'].unique())
```

```python
plt.figure(figsize=(14, 7))

sns.histplot(x="area_sqm", data=df, y="price", hue='flat_model', hue_order=flat_model_order)
```

```python
sns.jointplot(x='area_sqm', y='price', data=df, hue='flat_model', hue_order=flat_model_order)
```
