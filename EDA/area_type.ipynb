```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('train.csv')

df = df_original.copy()
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(x="area_sqm", y="price", data=df, hue='type', hue_order=df['type'].unique().sort())
```

```python
type_order = sorted(df['type'].unique())
```

```python
sns.jointplot(x='area_sqm', y='price', data=df, hue='type', hue_order=type_order)
```

```python
flat_model_1_2 = df[df['type'].isin(['1 ROOM', '2 ROOM'])]
flat_model_3 = df[df['type'].isin(['3 ROOM'])]
flat_model_4 = df[df['type'].isin(['4 ROOM'])]
flat_model_5 = df[df['type'].isin(['5 ROOM'])]
flat_model_Ex = df[df['type'].isin(['EXECUTIVE'])]
flat_model_mul = df[df['type'].isin(['MULTI-GENERATION'])]


```
