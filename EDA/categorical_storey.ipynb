```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('../train.csv')

df = df_original.copy()
```

```python
df.columns
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(x="storey_range", data=df, order=df['storey_range'].unique().sort())
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(x="storey_range", data=df, hue='type', hue_order=df['type'].unique().sort())
```

```python
plt.figure(figsize=(14, 7))
sns.countplot(x="storey_range", data=df, hue='flat_model', hue_order=df['flat_model'].unique().sort())
```
