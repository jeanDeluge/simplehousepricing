```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('../train.csv')

df = df_original.copy()
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(x="commence_date", y="price", data=df, hue='street', hue_order=df['street'].unique().sort())
```

```python
sns.jointplot(x='commence_date', y='price', data=df, hue='street')
```
